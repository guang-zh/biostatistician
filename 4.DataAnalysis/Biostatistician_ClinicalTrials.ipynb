```python
# Comparison for Pearson Correlation vs. Shwizer Wolff measures
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Generate x and epsilon
x = np.random.uniform(-5, 5, 100)
epsilon = np.random.uniform(-0.7, 0.7, 100)

# Calculate y
y = x**2 + epsilon

# Plot x vs y
plt.scatter(x, y, alpha=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x and y')
plt.show()

# Calculate Pearson correlation
correlation, _ = pearsonr(x, y)
print("Pearson correlation:", correlation)

def empirical_copula(X,Y):
    comparison_X = (X <= X[:, np.newaxis])
    comparison_Y = (Y <= Y[:, np.newaxis])
    preliminar_X = comparison_X.sum(axis=1)
    preliminar_Y = comparison_Y.sum(axis=1)

    n = len(X)
    sw = 0
    for i in range(n):
        sw += np.abs((comparison_X[i]*comparison_Y).sum(axis=1)/n - (preliminar_X[i]*preliminar_Y)/(n**2)).sum()
    sw = (12*sw)/(n**2-1)

    return sw

print("Shweizer Wolff correlation:", empirical_copula(x,y))
```
